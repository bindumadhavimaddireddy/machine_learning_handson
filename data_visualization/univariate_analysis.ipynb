## Box Plot

#### Box plot shows the five-number summary of data:
- 1. Minimum score,
  2. First(lower) quartile
  3. Median
  4. Third(upper) quartile
  5. Maximum score

#### **Diagram**

#### **Minimum Score**

- It is the **lowest value**, excluding outliers

- It is shown at the **end of bottom whisker**

#### **Lower Quartile**

- **25% of values** fall below the lower quartile value

- It is also known as the **first quartile**.

#### **Median**

- Median marks the **mid-point of the data**

<!-- - It is shown by the **line that divides the box into two parts** -->

- **50%. that is Half the scores are greater than or equal to this value and half are less**.

- It is sometimes known as the **second quartile**.

#### **Upper Quartile**

- **75% of the values fall below the upper quartile value**

- It is also known as the **third quartile**.

<!-- - So, **25% of data are above this value**. -->

#### **Maximum Score**

- It is the **highest value**, excluding outliers

- It is shown at the **end of upper whisker**.

#### **Whiskers**

- The upper and lower whiskers represent **values outside the middle 50%**

- That is, the **lower 25% of values** and the **upper 25% of values**.

#### **Interquartile Range (or IQR)**

- This is the box plot showing the **middle 50% of scores**

- It is the **range between the 25th and 75th percentile**.

- Minimum and maximum points are defined based on IQR


#### Outliers

- Their behavior is extremely different than overall data
- Box plot is mainly to find the outliers
- Ex. Let's say almost all students got the job within particular range of salary. but there are very very few students who got soo much salary which is completely different from other students salary range

#### Actual Calculation of min and max points

```python
min_point = max(overall_min, Q1-1.5*IQR)
max_point = min(overall_max, Q3+1.5*IQR)
```

#### Now, Let's plot a box plot to find the average typical earnings for a game

````python
plt.figure(figsize= (15,10))
sns.boxplot(x = df["Global_Sales"])
plt.yticks(fontsize=20)
plt.xlabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
````

#### Analysis of Global Sales based on Box Chart

- 50% of the data is above 7.8 and 50% of the data is below 7.8
- 25% of data is below 5.2
- 75% of data is above 5.2
- No outlines on the lower side ( sales don't go below 0 )
- 75% of sales are below 11 million dollars
- Maximum sales is 21million dollars
- There are few games that have exceptionally high sales which are outliners
